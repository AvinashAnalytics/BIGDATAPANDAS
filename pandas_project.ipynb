```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('scatter chart for total Vs tip',fontsize=15)
plt.show()
```

```python
import pandas as pd
tips.info()
```

```python
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('linechart for size vs total bill')
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('bar plot for day vs total bill')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('boxplot for day vs total bill')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('violine plot for day vs total bill')
plt.show()
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python

sns.kdeplot(tips['total_bill'],shade=True,color='green')
```

```python
sns.pairplot(tips)
```
